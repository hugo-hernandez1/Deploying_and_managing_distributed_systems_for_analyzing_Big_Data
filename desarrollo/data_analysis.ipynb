```python
import pandas as pd

# ── 1. Cargar datos ──────────────────────────────────────────────────────────
monolithic_monitor_data_175_ms = pd.read_csv('./Monolithic/data/175_ms/monitor_report.csv')
monolithic_monitor_data_175_ms['timestamp'] = pd.to_datetime(monolithic_monitor_data_175_ms['timestamp'])

air_quality_175_ms = pd.read_csv('./Monolithic/data/175_ms/air_quality_report.csv')
air_quality_175_ms['timestamp'] = pd.to_datetime(air_quality_175_ms['timestamp'])

metrics = [
    "latency_mean_us",
    "publication_throughput_sum_Bps",
    "subscription_throughput_sum_Bps",
]

# ── 2. Latencia y throughput ─────────────────────────────────────────────────
monolithic_monitor_data_175_ms_filtered = monolithic_monitor_data_175_ms[
    monolithic_monitor_data_175_ms["value"].notna()
    & (monolithic_monitor_data_175_ms["value"] != 0)
]

latency_throughput_175_ms = (
    monolithic_monitor_data_175_ms_filtered[monolithic_monitor_data_175_ms_filtered["metric"].isin(metrics)]
    .groupby(["topic", "metric"])["value"]
    .mean()
    .unstack()
    .reset_index()
)
latency_throughput_175_ms["publication_throughput_sum_KiBps"] = (
    latency_throughput_175_ms["publication_throughput_sum_Bps"] / 1024
)

latency_throughput_175_ms["subscription_throughput_sum_KiBps"] = (
    latency_throughput_175_ms["subscription_throughput_sum_Bps"] / 1024
)
latency_throughput_175_ms["throughput_total_mean_KiBps"] = (
    latency_throughput_175_ms["publication_throughput_sum_KiBps"].fillna(0)
    + latency_throughput_175_ms["subscription_throughput_sum_KiBps"].fillna(0)
)

print("=== Latencia y Throughput (175 ms) ===")
display(latency_throughput_175_ms)

# ── 3. Tiempos de reinicio (Opción A) ────────────────────────────────────────
restart_results = []

for (writer, topic), group in air_quality_175_ms[air_quality_175_ms['source_writer'].notna()].groupby(['source_writer', 'topic']):
    group = group.sort_values('timestamp')

    out_orders = group[group['event_type'] == 'OUT_OF_ORDER']
    if out_orders.empty:
        continue

    t_first_oo = out_orders['timestamp'].min()

    pre_samples = group[
        (group['event_type'] == 'SAMPLE') &
        (group['timestamp'] < t_first_oo)
    ]
    if pre_samples.empty:
        continue

    t_last_sample = pre_samples['timestamp'].max()
    restart_time_ms = (t_first_oo - t_last_sample).total_seconds() * 1000

    restart_results.append({
        'source_writer': writer,
        'topic': topic,
        'last_sample_before_restart': t_last_sample,
        'first_message_after_restart': t_first_oo,
        'restart_time_ms': restart_time_ms,
    })

restart_time_175_ms = pd.DataFrame(restart_results)

print("\n=== Tiempos de reinicio (175 ms) ===")
display(restart_time_175_ms)
```

```python
dist_monitor_data_250_ms_pub1 = pd.read_csv('./Cluster_pub1/data/250_ms/monitor_report.csv')
throughput_air_quality_raw_250_ms = dist_monitor_data_250_ms_pub1[
    (dist_monitor_data_250_ms_pub1["metric"] == "publication_throughput_sum_Bps") &
    (dist_monitor_data_250_ms_pub1["topic"] == "air_quality_raw")
]
throughput_air_quality_raw_250_ms.head()
```

```python
import pandas as pd

# ── Cargar el reporte del suscriptor final del cluster ──────────────────────
air_quality_175_ms = pd.read_csv('./Cluster_final/data/175_ms/air_quality_report.csv')
air_quality_175_ms['timestamp'] = pd.to_datetime(air_quality_175_ms['timestamp'])

# ── Tiempo de recuperación por pod (publisher) ──────────────────────────────
#
# Para cada source_writer (cada pod publicador):
#   - Detectamos el primer OUT_OF_ORDER → momento en que el pod se reinició
#     y el suscriptor empieza a recibir mensajes con sequence_id reseteado.
#   - El último SAMPLE antes de ese momento es la última muestra recibida
#     antes de la caída.
#   - restart_time = primer_OUT_OF_ORDER - último_SAMPLE_antes

restart_results = []

for (writer, topic), group in air_quality_175_ms[
    air_quality_175_ms['source_writer'].notna()
].groupby(['source_writer', 'topic']):
    group = group.sort_values('timestamp')

    out_orders = group[group['event_type'] == 'OUT_OF_ORDER']
    if out_orders.empty:
        continue

    t_first_oo = out_orders['timestamp'].min()

    pre_samples = group[
        (group['event_type'] == 'SAMPLE') &
        (group['timestamp'] < t_first_oo)
    ]
    if pre_samples.empty:
        continue

    t_last_sample = pre_samples['timestamp'].max()
    restart_time_ms = (t_first_oo - t_last_sample).total_seconds() * 1000

    restart_results.append({
        'source_writer': writer,
        'topic': topic,
        'last_sample_before_restart': t_last_sample,
        'first_message_after_restart': t_first_oo,
        'restart_time_ms': restart_time_ms,
    })

restart_time_175_ms = pd.DataFrame(restart_results)
restart_time_175_ms
```
